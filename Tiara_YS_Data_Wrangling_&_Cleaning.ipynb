## Data Formatting

Delete unecessary format in the data

```python
home2['luas_rumah'] = home2['house_area'].str.replace(' ', '', regex=False)
home2
```

```python
home2['luas_rumah'] = home2['house_area'].str.extract(r'(.*?)(?=m)', expand=False).astype(float) #.str.extract(r'(\d+\.\d+)') #.apply(lambda x: ''.join(re.findall(r'\d.', x)))
home2
```

```python
home2['harga_per_m2'] = home2['price_area'].str.replace(r'Rp\s*|\s*per\s*mÂ²', '', regex=True)
home2
```

```python
home2['harga_per_m2'] = home2['harga_per_m2'].str.replace('.', '', regex=False).astype(float)
home2
```

```python
home2['satuan_harga'] = home2['price'].str.replace(r'[^A-Za-z]', '', regex=True)
home2
```

```python
home2['satuan_harga'].unique()
```

```python
home2['satuan_harga'].value_counts()
```

```python
#ganti satuan
def kategori_harga(value):
    if "jtRpM" in value:
      return "Juta"
    elif "MRpM" in value:
      return "Milyar"
    if "M" in value:
      return "Milyar"
    elif "jt" in value:
      return "Juta"
    return "Juta"

home2['kategori_harga'] = np.vectorize(kategori_harga)(home2["satuan_harga"])
home2
```

```python
home2['harga_bangunan'] = (
    home2['price']                       # kolom asal
      .astype(str)                    # jagaâ€‘jaga kalau ada tipe nonâ€‘string
      .str.extract(r'^\s*([0-9]+(?:[.,][0-9]+)?)', expand=False)  # ambil angka pertama
      .str.replace(',', '.', regex=False)  # ubah koma â†’ titik kalau ada
      .astype(float)                  # opsional: jadi numeric
)
home2
```

```python
home2[home2['harga_bangunan'].astype(str).str.count(r'\.') > 1]
```

```python
def konvert_harga(harga, kategori):
    if isinstance(harga, (float, int)):
        return harga * (1000000000 if kategori == 'Milyar' else 1000000)
    else:
        return pd.to_numeric(harga.replace('.', '.'), errors='coerce') * (1000000000 if kategori == 'Milyar' else 1000000)

home2['harga_rumah'] = home2.apply(lambda row: konvert_harga(row['harga_bangunan'], row['kategori_harga']), axis=1)
pd.set_option('display.float_format', '{:.2f}'.format)
home2
```

```python
home2[~home2['listing_location'].str.contains(r'Tangerang\s+Selatan',
                                                 case=False,      # abaikan kapital/kecil
                                                 na=False,        # anggap NaN sebagai 'tidak mengandung'
                                                 regex=True)]
# semua rumah ada di tangerang selatan
```

```python
home2['lokasi'] = home2['listing_location'].str.split(',', n=1).str[0].str.strip()
home2
```

```python
# 1. Harga rumah â†’ mean per lokasi (round 2)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
mean_per_loc = (
    home3.groupby('lokasi')['harga_per_m2']
      .transform(lambda x: round(x.mean(), 2))   # hitung mean & bulatkan 2 desimal
)

home3['harga_tiap_m2'] = home3['harga_per_m2'].fillna(mean_per_loc)
```

```python
mean_floor_per_loc = (
    home3.groupby('lokasi')['bed']
      .transform(lambda x: np.floor(x.mean()).astype(int))
)

home3['jumlah_kamar'] = home3['bed'].fillna(mean_floor_per_loc)
```

```python
mean_floor_per_loc = (
    home3.groupby('lokasi')['bath']
      .transform(lambda x: np.floor(x.mean()).astype(int))
)

home3['jumlah_km'] = home3['bath'].fillna(mean_floor_per_loc)
```

```python
# Menambahkan data kecamatan dan kelurahan
kecamatan_list = ['Ciputat', 'Ciputat Timur', 'Pamulang', 'Pondok Aren', 'Serpong', 'Serpong Utara', 'Setu']
kelurahan_list = kelurahan_list = [
    "Cipayung", "Ciputat", "Jombang", "Sawah Baru", "Sawah Lama", "Serua", "Serua Indah",
    "Cempaka Putih", "Cireundeu", "Pisangan", "Pondok Ranji", "Rempoa", "Rengas",
    "Bambu Apus", "Benda Baru", "Kedaung", "Pamulang Barat", "Pamulang Timur",
    "Pondok Benda", "Pondok Cabe Ilir", "Pondok Cabe Udik",
    "Jurangmangu Barat", "Jurangmangu Timur", "Pondok Kacang Barat", "Pondok Kacang Timur",
    "Perigi Lama", "Perigi Baru", "Pondok Aren", "Pondok Karya", "Pondok Jaya",
    "Pondok Betung", "Pondok Pucung",
    "Buaran", "Ciater", "Cilenggang", "Lengkong Gudang", "Lengkong Gudang Timur",
    "Lengkong Wetan", "Rawa Buntu", "Rawa Mekar Jaya", "Serpong",
    "Jelupang", "Lengkong Karya", "Pakualam", "Pakulonan", "Paku Jaya",
    "Pondok Jagung", "Pondok Jagung Timur",
    "Babakan", "Bakti Jaya", "Kademangan", "Keranggan", "Muncul", "Setu"
]
```

```python
def extract_kecamatan(listing_location, kecamatan_list):
    for kecamatan in kecamatan_list:
        if kecamatan.lower() in listing_location.lower():
            return kecamatan
    return None

home4["Kecamatan"] = home4["lokasi"].apply(lambda x: extract_kecamatan(x, kecamatan_list))
```

```python
def extract_kelurahan(listing_location, kelurahan_list):
    for kelurahan in kelurahan_list:
        if kelurahan.lower() in listing_location.lower():
            return kelurahan
    return None

home4["kelurahan"] = home4["lokasi"].apply(lambda x: extract_kelurahan(x, kelurahan_list))
```
